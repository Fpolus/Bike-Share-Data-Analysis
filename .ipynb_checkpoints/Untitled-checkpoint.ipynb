```python
import pandas as pd
```

```python
bike_data = pd.read_csv('resources/NYC-CitiBike-2016.csv')
bike_data
```

```python
bike_data.info()
```

```python
# separating the date and time

# first let's separate the start time and date
bike_data['date'] = pd.to_datetime(bike_data['starttime']).dt.date
bike_data['time'] = pd.to_datetime(bike_data['starttime']).dt.time

# Now we are going to separate the end time and date
bike_data['date'] = pd.to_datetime(bike_data['stoptime']).dt.date
bike_data['time'] = pd.to_datetime(bike_data['starttime']).dt.time
```
